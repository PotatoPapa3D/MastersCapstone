# Splitting the data into training and test sets

```python
# Setting my feature and response data
X = data.loc[:, data.columns != 'default payment next month']
y = data.loc[:, data.columns == 'default payment next month']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
```

# Checking for imbalanced data and balancing the data

```python
# I need to check and see how balanced my training response data is, this is located in the column titled "default payment next month"
y_train.value_counts()
```

```python
# I am going to visualize the imbalance with a pie chart, below it shows that the data is very imbalanced. 
y_train.value_counts().plot.pie(autopct='%.2f')
```

```python
# I am going to manually random-undersample my data in order to balance it. This means that I am going to randomly select
# 5009 rows of data where my response variable is 0 so that there will be an equal representation of 0's and 1's in my response
# data. Step 1- I am going to start by creating a variable equal to the number of 1's that I have.
mino_class_len = 0
for i in y_train['default payment next month']:
    if i == 1:
        mino_class_len += 1
    else:
        continue
print(mino_class_len)
```

```python
# Step 2- assigning all indices values where our response variable equals 0 to another variable.
majo_class_indices = y_train.index[y_train['default payment next month'] == 0]
```

```python
# Step 3- randomly selecting 5009 (the number of 1's I have) rows of data where my response variable is equal to 0.
random_majo_indices = np.random.choice(majo_class_indices, mino_class_len, replace = False)
print(len(random_majo_indices))
```

```python
# Step 4- assigning all indices values where our response variable equals 1 to another variable.
mino_class_indices = y_train.index[y_train['default payment next month'] == 1]
```

```python
# Step 5- Concatenating the indices from step 3 and step 4 to create a list of indices.
under_samp_indices = np.concatenate([mino_class_indices, random_majo_indices])
```

```python
# Step 6- Selecting all rows of data from our original data that have the same indices that we created from undersampling.
un_samp = data.loc[under_samp_indices]
```

```python
# Step 7- Assigning X and y or feature and response variables.
X_train = un_samp.loc[:, un_samp.columns != 'default payment next month']
y_train = un_samp.loc[:, un_samp.columns == 'default payment next month']
```

```python
# Step 8- Making sure the process above balanced our data.
y_train.value_counts()
```

```python
# Step 9- Visualize our new response data.
y_train.value_counts().plot.pie(autopct='%.2f')
```

# Scaling of Features

```python
# I am using the MinMax Scaler to scale the data.
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings("ignore")

scaler = MinMaxScaler()
scaler.fit(X_train.iloc[:, 11:])

X_train.iloc[:, 11:] = scaler.transform(X_train.iloc[:, 11:])
X_test.iloc[:, 11:] = scaler.transform(X_test.iloc[:, 11:])
```

# KNN Classifier

```python
# I am going to start by training a KNN Classifier model. Here I am importing the model and running a basic 
# cross validation score on the model with n_neighbors being set to five and the scoring for the cross validation being 
# set to accuracy. 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=5)

scores = cross_val_score(knn, X, y.values.ravel(), cv=10, scoring='accuracy')
print(scores)
```

```python
# To find the right number of neighbors I am going to have a for loop test each number of neighbors for 1 to 15 in this 
# block of code and then 15 to 30 in the next block. The most accurate neighbor will be used to train the final KNN model.
#num_neigh = range(1, 15, 1)

#cv_scores = []

#for n in num_neigh:
#    knn = KNeighborsClassifier(n_neighbors=n)
#    scores = cross_val_score(knn, X, y.values.ravel(), cv=10, scoring='accuracy')
#    cv_scores.append(scores.mean())
    
#cv_scores_df = pd.DataFrame({"N": [n for n in num_neigh], 
#              "Accuracy": cv_scores}).set_index("N")
#print(cv_scores_df)
```

```python
num_neigh = range(15, 30, 1)

cv_scores = []

for n in num_neigh:
    knn = KNeighborsClassifier(n_neighbors=n)
    scores = cross_val_score(knn, X, y.values.ravel(), cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
    
cv_scores_df = pd.DataFrame({"N": [n for n in num_neigh], 
              "Accuracy": cv_scores}).set_index("N")
print(cv_scores_df)
```

```python
# I am creating a variable equal to the n_neighbor with the highest accuracy score. 
best_neighbor = cv_scores_df['Accuracy'].idxmax()
best_neighbor
```

```python
# Fitting and predicting y values from my test set. 
knn2 = KNeighborsClassifier(n_neighbors=best_neighbor)
knn2.fit(X_train, y_train.values.ravel())
y_pred = knn2.predict(X_test)
```

```python
# Importing my evaluation metrics.
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, roc_auc_score
```

```python
# Creating a function that takes he name of the model in and the predicted y values and then prints and returns the 
# respective metrics for that model. I will use the returned values later on for graphing purposes.
def scores(model_name, y_pred_name):
    acc = round(accuracy_score(y_test.values.ravel(), y_pred_name), 2)
    f1 = round(f1_score(y_test.values.ravel(), y_pred_name), 2)
    rc = round(recall_score(y_test.values.ravel(), y_pred_name), 2)
    ps = round(precision_score(y_test.values.ravel(), y_pred_name), 2)
    ra = round(roc_auc_score(y_test.values.ravel(), y_pred_name), 2)
    print("Accuracy Score: ", acc,
          "\nF1 Score: ", f1,
          "\nRecall Score: ", rc,
          "\nPrecision Score: ", ps,
          "\nRoc Auc Score: ", ra)
    return model_name, acc, f1, rc, ps, ra
```

```python
KNN_Scores = scores('KNN', y_pred)
KNN_Scores
```
